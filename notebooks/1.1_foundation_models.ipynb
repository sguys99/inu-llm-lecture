### 2. init_chat_model을 활용한 모델 호출

```python
pprint(response.content)
```

### 3. 대화이력 입력하기

```python
from langchain.messages import HumanMessage, AIMessage
```

```python
response = llm.invoke(
    input= [HumanMessage(content="달의 수도는 어디야?")] # 메시지가 사용자로 부터 왔음을 표시
)
response
```

대화이력을 함께 입력해줘.

```python
response = llm.invoke(
    input= [HumanMessage(content="달의 수도는 어디야?"),
    AIMessage(content="달의 수도는 Luna City입니다."),
    HumanMessage(content="재미있네요, Luna City에 대해서 더 알려주세요.")]
    )

pprint(response)
```

### 4. 에이전트 생성, 호출하기

```python
from langchain.agents import create_agent
```

```python
agent = create_agent(model="claude-haiku-4-5")
```

```python
response = agent.invoke(
    {"messages": [HumanMessage(content="달의 수도는 어디야?")]} # 메시지가 담긴 딕셔너리를 보내고, 그것이 사용자로 부터 왔음을 표시
)
pprint(response)
```

마지막 메시지만 추출

```python
print(response['messages'][-1].content)
```

대화이력 입력하기

```python
response = agent.invoke(
    {"messages": [HumanMessage(content="달의 수도는 어디야?"),
    AIMessage(content="달의 수도는 Luna City입니다."),
    HumanMessage(content="재미있네요, Luna City에 대해서 더 알려주세요.")]}
)

pprint(response)
```

### 5. 스트리밍 출력하기

```python
for token, metadata in agent.stream(
    {"messages": [HumanMessage(content="달의 수도인 Luna City에 대해서 알려주세요")]},
    stream_mode="messages"
):

    # token is a message chunk with token content
    # metadata contains which node produced the token
    
    if token.content:  # Check if there's actual content
        print(token.content, end="", flush=True)  # Print token
```

스트리밍 모드 종류
- "values": 그래프의 전체 상태를 스트리밍
- "updates": 각 노드의 업데이트만 스트리밍
- "messages": 메시지 청크(토큰)를 실시간으로 스트리밍 ✅
